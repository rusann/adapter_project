## Желательные составляющие решения и система оценки

### Часть 1: Исследование блоков (4 балла)

На лекции мы смотрели на `up_blocks.0` и `down_blocks.2 + mid_block` как на два полюса. Здесь нужно разобраться детальнее.

- Проведите полный ablation: переберите **каждый блок по отдельности** (`up_blocks.0`, `up_blocks.1`, `down_blocks.2`, `down_blocks.1`, `mid_block`), зафиксируйте один референс, один промпт, один seed. Результаты в grid с подписями — 1 балл
- Найдите **минимальную комбинацию блоков** при которой стиль ещё читается — то есть уберите всё лишнее. Обоснуйте выбор письменно — 1 балл
- Проверьте найденную комбинацию на **трёх разных промптах** и **трёх разных референсах** — насколько она устойчива? Есть ли случаи где она ломается? — 1 балл

### Часть 2: Два стиля одновременно (3 балла)

IP-Adapter принимает image embedding — это просто тензор. Значит можно интерполировать между двумя стилями в пространстве эмбеддингов.

- Реализуйте интерполяцию двух эмбеддингов и сделайте sweep по alpha, результат в grid — 1 балл
- Попробуйте **нелинейные смешивания** — например взять норму каждого эмбеддинга в расчёт, или смешивать только часть токенов (первые 8 из 16 от стиля A, вторые 8 от стиля B). Опишите что попробовали и что получилось — 1 балл
- Примените интерполяцию **совместно с block-selective иньекцией** из части 1: стиль A только через стилевые блоки, стиль B только через контекстные. Сравните с обычной интерполяцией — 1 балл

### Часть 3: Реальный пайплайн (3 балла)

Совместите IP-Adapter с ControlNet Canny — структура берётся из одного изображения, стиль из другого.

- Соберите пайплайн `StableDiffusionXLControlNetPipeline` + IP-Adapter, убедитесь что оба работают одновременно — 1 балл
- Сделайте grid 3×3: по одной оси `controlnet_scale` [0.3, 0.7, 1.0], по другой `ip_adapter_scale` [0.3, 0.7, 1.0], один и тот же промпт/структура/стиль — 1 балл
- Примените block-selective инжекцию из части 1 внутри этого пайплайна. Сравните: обычный IP-Adapter + ControlNet vs InstantStyle + ControlNet — визуально что лучше сохраняет структуру? — 1 балл
- Письменно: опишите trade-off между тремя силами — промпт, структура, стиль. При каких значениях что доминирует? Где оптимум на ваш взгляд? — 1 балл
